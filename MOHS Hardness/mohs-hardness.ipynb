```python
train_data = pd.read_csv('/kaggle/input/playground-series-s3e25/train.csv')
train_data
```

```python
test_data = pd.read_csv('/kaggle/input/playground-series-s3e25/test.csv')
test_data
```

```python
features_ = ['allelectrons_Total', 'density_Total', 'allelectrons_Average',
       'val_e_Average', 'atomicweight_Average', 'ionenergy_Average',
       'el_neg_chi_Average', 'R_vdw_element_Average', 'R_cov_element_Average',
       'zaratio_Average', 'density_Average']
x_train = np.array(train_data[features_])
x_test = np.array(test_data[features_])
x_train
x_test
```

```python
y_train = np.array(train_data['Hardness'])
y_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train,y_train)
predictions = model.predict(x_test)
predictions
```

```python
df = pd.DataFrame({'id':test_data['id'],'Hardness':predictions})
df
```

```python
df.to_csv('submission.csv',index=True)
```
